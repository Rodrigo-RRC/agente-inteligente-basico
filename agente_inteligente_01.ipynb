#Explicação

- O transformers é da Hugging Face e permite que a gente use modelos de linguagem já treinados, como o GPT-2.
- O torch é o PyTorch, a base que esses modelos usam para funcionar.


#Carregar o modelo e fazer o agente gerar uma resposta

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Escolher um modelo de linguagem pronto para gerar texto
modelo_nome = "gpt2"

# Carregar o tokenizer (quebra o texto em partes que o modelo entende)
tokenizer = AutoTokenizer.from_pretrained(modelo_nome)

# Carregar o modelo pré-treinado
modelo = AutoModelForCausalLM.from_pretrained(modelo_nome)

# Criar a entrada para o modelo
entrada = "Olá, tudo bem? Me fale uma curiosidade."

inputs = tokenizer(entrada, return_tensors="pt")

# Gerar uma resposta com o modelo
outputs = modelo.generate(**inputs, max_length=100, do_sample=True, temperature=0.9)

# Mostrar a resposta gerada
resposta = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(resposta)
```
